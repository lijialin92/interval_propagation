```python
# fifo-list with counter example, split total time are[10min, 20min, 30min, 40min, 50min, 60min]
run_times = [600, 1200, 1800, 2400, 3000, 3600]

# generate the input interval
input_interval_min = np.array([[10., 10.]])
input_interval_max = np.array([[50., 50.]])
input_interval = [input_interval_min, input_interval_max]

for run_time in run_times:
    print("****************************FIFO list with finding counter example, the run time for split is " + str(run_time) + ".*********************************************************")
    start_time = time.time()
    res = model_3_certification_with_single_input_interval_fifo(input_interval, run_time)
    used_time = time.time() - start_time
    coverage = model_3_calculate_coverage(res)
    print("satisfication: " + str(res["sat"]))
    print("number of safe intervals: " + str(len(res["safety_intervals"])))
    print("number of unsafe intervals: " + str(len(res["counter_intervals"])))
    print("number of counter examples: " + str(len(res["counter_examples"])))
    print("number of uncertain intervals: " + str(len(res["uncertain_intervals"])))
    print("coverage: " + str(coverage))
    print("certification used time: " + str(used_time))
```
