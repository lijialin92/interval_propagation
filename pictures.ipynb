```python
import matplotlib
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import ticker
%matplotlib inline

plt.figure()
matplotlib.rc('axes', facecolor = 'white')
matplotlib.rc('figure', figsize = (8,5))
matplotlib.rc('font', size = 14)
matplotlib.rc('axes', grid = True)

plt.xlabel('L infinity perturbation')
plt.ylabel('Certified robustness')

x = [0.5, 1.0, 1.5, 2.0, 2.5, 3.0]
y1 = [100, 100, 99.8, 99.5, 99.1, 98.6]#large pri, search large
y2 = [100, 100, 99.8, 99.5, 99.1, 98.6]#small pri, search large
y3 = [99.9, 99.5, 97.6, 83.7, 71.1, 62.2]#plain interval propagation
plt.plot(x, y1, color = 'r', marker = 'o')
plt.plot(x, y2, color = 'b', marker = 'v')
plt.plot(x, y3, color = 'g', marker = 's')
plt.legend(['type 1', 'type 2', 'type 3'], loc='lower center')

plt.yticks([100,80,60,40,20,0],
            ["100%", "80%", "60%", "40%", "20%", "0%"])
plt.savefig('l_infinity_perturbation_large_small_pri.png', dpi=300, bbox_inches = 'tight')
```
