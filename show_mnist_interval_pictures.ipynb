```python
# plot the picture after change the value of the pixels of the pictures. because we want to make sure that after change, people(or we say human) should be possible to recogenize the picture. 
# eg. train_x[0] is 5, we add each pixel 5(maximum 255). After adding 5 to each pixel, we show the picture can be still recogenized be 5.
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist

(x_train, y_train), (x_test, y_test) = mnist.load_data()
print(x_train.shape)
print(y_train[0])

# pick a sample to plot
sample = 1
image = x_train[0]# plot the sample
fig = plt.figure()
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.show()

# pixel plus 5
image_5 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 250:
            image_5[row][col] = image[row][col] + 5
        else:
            image_5[row][col] = 255

fig5 = plt.figure()
plt.imshow(image_5, cmap='gray', vmin=0, vmax=255)
plt.show()

# pixel plus 55
image_55 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 200:
            image_55[row][col] = image[row][col] + 55
        else:
            image_55[row][col] = 255

fig55 = plt.figure()
plt.imshow(image_55, cmap='gray', vmin=0, vmax=255)
plt.show()

# pixel plus 155
image_155 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 100:
            image_155[row][col] = image[row][col] + 155
        else:
            image_155[row][col] = 255
fig155 = plt.figure()
plt.imshow(image_155, cmap='gray',vmin=0, vmax=255)
plt.show()
```

```python
# where the pixel value is 0, then still remain in 0. where is more than 0 then plus 5, 55, 155
sample = 1
image = x_train[0]# plot the sample
fig = plt.figure()
plt.imshow(image, cmap='gray', vmin=0, vmax=255)
plt.savefig('x_train_0.png')
plt.show()

# pixel plus 5
image_5 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 250 and image[row][col] != 0:
            image_5[row][col] = image[row][col] + 5
        elif image[row][col] > 250:
            image_5[row][col] = 255

fig5 = plt.figure()
plt.imshow(image_5, cmap='gray', vmin=0, vmax=255)
plt.savefig('x_train_0_5.png')
plt.show()

# pixel plus 55
image_55 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 200 and image[row][col] != 0:
            image_55[row][col] = image[row][col] + 55
        elif image[row][col] > 200:
            image_55[row][col] = 255

fig55 = plt.figure()
plt.imshow(image_55, cmap='gray', vmin=0, vmax=255)
plt.savefig('x_train_0_55.png')
plt.show()

# pixel plus 155
image_155 = image.copy()
for row in range(image.shape[0]):
    for col in range(image.shape[1]):
        if image[row][col] <= 100 and image[row][col] != 0:
            image_155[row][col] = image[row][col] + 155
        elif image[row][col] > 100:
            image_155[row][col] = 255

fig155 = plt.figure()
plt.imshow(image_155, cmap='gray', vmin=0, vmax=255)
plt.savefig('x_train_0_155.png')
plt.show()
```
